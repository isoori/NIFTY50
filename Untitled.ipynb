```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Final-50-stocks.csv')
df.head()
```

```python
dff = df.copy()
```

```python
for i in range(1,df.shape[0]):
    for j in range(1,df.shape[1]):
        dff.iloc[i,j] = df.iloc[i,j] - df.iloc[i-1,j]
```

```python
dff.drop(0,axis=0,inplace=True)
```

```python
dff
```

```python
stacks = dff.columns.tolist()
```

```python
new_df = pd.DataFrame()
```

```python
for col in stacks :
    if col =='DATE':
        continue
    new_df.loc[0,col] = np.abs(dff[col]).sum(axis=0) / dff[col].notna().sum()
```
